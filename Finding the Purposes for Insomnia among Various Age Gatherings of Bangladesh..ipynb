```python
import matplotlib.pyplot as plt
import numpy as np
w=0.3
x = ["13-19","20-25","26-36","37-50"," above 50"]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]

male=[0,20,50,0,0]
female=[33.33,80,75,50,25]

plt.bar(bar1,male,w,label="male")
plt.bar(bar2,female,w,label="female")

plt.xlabel("Age groups")
plt.ylabel("Percentage(%)")

plt.legend()

plt.xticks(bar1+w,x)




plt.savefig('insomniafig\daytimesleepfig.pdf',bbox_inches='tight')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
w=0.4
x = ["13-19","20-25","26-36","37-50"," above 50"]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]

fear=[50,0,0,0,0]
overthinking=[75,40,83.33,40,60]
depression=[0,0,50,40,60]
anxiety=[0,0,0,40,60]
none=[0,40,0,0,0]




b_an=list(np.add(anxiety,overthinking))
b_non=list(np.add(fear,anxiety))
b_no=list(np.add(fear,none))


plt.bar(x,overthinking,w,label="overthinking")
plt.bar(x,depression,w,bottom=overthinking,label="depression")
plt.bar(x,anxiety,w,bottom=b_an,label="anxiety")
plt.bar(x,fear,w,bottom=b_non,label="fear")
plt.bar(x,none,w,bottom=b_no,label="no pressure")




plt.xlabel("Age groups")
plt.ylabel("Percentage(%)")

plt.legend()

plt.xticks(bar1,x)



plt.savefig('insomniafig\meantalpressure.pdf',bbox_inches='tight')

plt.show()

```
